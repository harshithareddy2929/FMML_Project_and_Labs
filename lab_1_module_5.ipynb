#QUESTION 1)Try to run the same experiment after filtering different features, 2 at a time.

#ANSWER 1)

```python
from sklearn.datasets import load_iris
from sklearn import svm
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)


ar = np.vstack([
    np.random.multivariate_normal(np.array([1, 1]), 1.5 * np.array([[2, -1], [-1, 2.0]]), size=50),
    np.random.multivariate_normal(np.array([3, 3]), 2 * np.array([[0.75, -0.5], [-0.5, 0.75]]), size=50),
])

y = np.array([0] * int((ar.shape[0] / 2)) + [1] * int((ar.shape[0] / 2)))


def plotDecisionBoundary(X, y, pair, clf):
    x_min, x_max = X[:, pair[0]].min() - 1, X[:, pair[0]].max() + 1
    y_min, y_max = X[:, pair[1]].min() - 1, X[:, pair[1]].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                         np.arange(y_min, y_max, 0.1))

    y_pred = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    y_pred = y_pred.reshape(xx.shape)
    plt.figure(figsize=(5, 4))
    plt.contourf(xx, yy, y_pred, alpha=0.4)
    plt.scatter(X[:, pair[0]], X[:, pair[1]], c=y, s=50, edgecolor='k')
    plt.gcf().set_dpi(130)
    plt.show()

def boundaryExp(X, y, pair):
    clf = svm.LinearSVC()
    clf.fit(X[:, pair], y)
    plotDecisionBoundary(X, y, pair, clf)
feature_pairs = [[0, 1]]

for pair in feature_pairs:
    boundaryExp(ar, y, pair)
```

#QUESTION 2)Using different kinds of kernels for the SVM among: {‘linear’, ‘poly’, ‘rbf’, ‘sigmoid’, ‘precomputed’}

#ANSWER 2)

```python

from sklearn.datasets import load_iris, make_circles
from sklearn import svm
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go

np.random.seed(0)

ar = np.vstack([
    np.random.multivariate_normal(np.array([1, 1]), 1.5 * np.array([[2, -1], [-1, 2.0]]), size=50),
    np.random.multivariate_normal(np.array([3, 3]), 2 * np.array([[0.75, -0.5], [-0.5, 0.75]]), size=50),
])

y = np.array([0] * int((ar.shape[0] / 2)) + [1] * int((ar.shape[0] / 2)))

def plotDecisionBoundary(X, y, pair, clf):
    x_min, x_max = X[:, pair[0]].min() - 1, X[:, pair[0]].max() + 1
    y_min, y_max = X[:, pair[1]].min() - 1, X[:, pair[1]].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                         np.arange(y_min, y_max, 0.1))

    y_pred = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    y_pred = y_pred.reshape(xx.shape)
    plt.figure(figsize=(5, 4))
    plt.contourf(xx, yy, y_pred, alpha=0.4)
    plt.scatter(X[:, pair[0]], X[:, pair[1]], c=y, s=50, edgecolor='k')
    plt.gcf().set_dpi(130)
    plt.show()

def kernelExp(X, y, pair, kernel_type):
    clf = svm.SVC(kernel=kernel_type)
    clf.fit(X[:, pair], y)
    plotDecisionBoundary(X, y, pair, clf)

feature_pairs = [[0, 1]]
kernel_types = ['linear', 'poly', 'rbf', 'sigmoid']

for kernel_type in kernel_types:
    print(f"Running experiment with {kernel_type} kernel:")
    kernelExp(ar, y, feature_pairs[0], kernel_type)
```
