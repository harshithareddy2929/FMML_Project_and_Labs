```python
!pip install wikipedia
!pip install plotly

import wikipedia
import nltk
from nltk.util import ngrams
from collections import Counter
import matplotlib.pyplot as plt
import re
import plotly.express as px
import pandas as pd

# Set the language for Wikipedia
wikipedia.set_lang('en')

topic1 = 'Giraffe'
topic2 = 'Elephant'

eng1 = wikipedia.page(topic1).content
eng2 = wikipedia.page(topic2).content

wikipedia.set_lang('fr')

fr1 = wikipedia.page(topic1).content
fr2 = wikipedia.page(topic2).content

def cleanup(text):
    text = text.lower()  # Make it lowercase
    text = re.sub('[^a-z]+', ' ', text)  # Only keep alphabetic characters
    return text

eng1 = cleanup(eng1)
eng2 = cleanup(eng2)
fr1 = cleanup(fr1)
fr2 = cleanup(fr2)

# Function to plot n-gram histograms
def plot_ngram_histogram(ngram_counter, title):
    ngram_df = pd.DataFrame(ngram_counter.items(), columns=['N-gram', 'Frequency'])
    ngram_df = ngram_df.sort_values(by='Frequency', ascending=False)

    fig = px.bar(ngram_df, x='N-gram', y='Frequency', title=title)
    fig.show()

# convert a tuple of characters to a string
def tuple2string(tup):
    st = ''
    for ii in tup:
        st = st + ii
    return st

# convert a tuple of tuples to a list of strings
def key2string(keys):
    return [tuple2string(i) for i in keys]

# Trigrams
trigram_eng1 = Counter(ngrams(eng1.split(), 3))  # Trigrams
plot_ngram_histogram(trigram_eng1, 'English 1 - Trigrams')

trigram_eng2 = Counter(ngrams(eng2.split(), 3))  # Trigrams
plot_ngram_histogram(trigram_eng2, 'English 2 - Trigrams')

trigram_fr1 = Counter(ngrams(fr1.split(), 3))  # Trigrams
plot_ngram_histogram(trigram_fr1, 'French 1 - Trigrams')

trigram_fr2 = Counter(ngrams(fr2.split(), 3))  # Trigrams
plot_ngram_histogram(trigram_fr2, 'French 2 - Trigrams')

# You can similarly compute and plot higher n-grams (e.g., 4-grams) by changing the n value in ngrams().

# 4-grams
fourgram_eng1 = Counter(ngrams(eng1.split(), 4))  # 4-grams
plot_ngram_histogram(fourgram_eng1, 'English 1 - 4-grams')

fourgram_eng2 = Counter(ngrams(eng2.split(), 4))  # 4-grams
plot_ngram_histogram(fourgram_eng2, 'English 2 - 4-grams')

fourgram_fr1 = Counter(ngrams(fr1.split(), 4))  # 4-grams
plot_ngram_histogram(fourgram_fr1, 'French 1 - 4-grams')

fourgram_fr2 = Counter(ngrams(fr2.split(), 4))  # 4-grams
plot_ngram_histogram(fourgram_fr2, 'French 2 - 4-grams')

# You can continue this pattern for higher n-grams as needed.


```

When we are dealing with more than two classes in a classification task, it becomes a multi-class classification problem. In the context of digit recognition using the MNIST dataset, this means we are trying to classify each digit into one of the ten possible classes (0 through 9).

Here's what happens when we work with multi-class classification:

1.Class Labels: Each digit image in the dataset is associated with a specific class label (0 through 9) that represents the true identity of the digit.

2.Model Output: Our machine learning model is trained to output a probability distribution over all possible classes for each input image. In other words, instead of just predicting "0" or "1," the model now predicts probabilities for all ten classes.

3.Loss Function: We typically use a multi-class loss function, such as categorical cross-entropy, to train the model. This loss function measures the dissimilarity between the predicted probability distribution and the true class labels.

4.Evaluation Metrics: We evaluate the model's performance using metrics like accuracy, precision, recall, F1-score, and confusion matrices for multi-class classification. These metrics help you assess how well the model is doing across all classes.

5.Softmax Activation: The final layer of our neural network often uses a softmax activation function, which ensures that the predicted probabilities sum to 1 across all classes. This is used to convert the model's raw output into a probability distribution.

6.One-Hot Encoding: In multi-class classification, the true class labels are typically one-hot encoded. Each label is represented as a binary vector, where only one element is 1 (indicating the true class), and all other elements are 0.

7.Decision Rule: To make a prediction, we select the class with the highest predicted probability as the predicted label for a given input.

8.Handling Imbalanced Data: We may need to consider strategies to address class imbalance if some classes have significantly fewer examples than others.

In summary, when working with more than two classes, we need to adapt our model architecture, loss function, and evaluation metrics to handle multi-class classification. The model learns to distinguish between all possible classes simultaneously, and its output consists of probability distributions over those classes. The choice of features, model architecture, and training strategy should be designed to effectively capture and discriminate between the characteristics of all classes in our dataset.
